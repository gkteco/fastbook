# Fast iteration is key

Don't be a perfectionist. It's not about finding the perfect project, perfect dataset, or knowoing "everything" before getting started. It's about trying things, iterating, and *actively* learning. You could be on your **third** iteration before the perfectionist even start.

## complete projects end-to-end

Don't spend all your time working on fine-tuning a model or labeling a dataset when you're in learning/protyping mode. Each iteration of the project should yield the end result you are after. The benifits with this approach is:

1) You get a high level overview of the architecture and the project. This could reveal where the "gotchas" are and where time needs to be spent as you complete the project.
2) You'll get a sense for what is "good enough". Why eek out another 2% of accuracy out of the model if it is doing the job. For example, if the end result is to develop an AI powered mobile application, all early stage prototyping could be done using a simple web app.

## You need to understand the current state of AI/ML. 

You should not attempt a prototype of something that has not been done before. Documentation is extremly important and finding code/libraries that have sloved the problem you are trying to solve. If you embark on a journey where there is no examples or documedtation then you could fail at completing the prototype.

Stay mindful for what can and can't be done in the current state of AI/ML. It is more harmful to say something can be done (when it can't) than to set expectations and shift the conversation. For example, if a customer want's a POC of a fully autonomous self driving car, this is something that cannot be done. However, let's drive the conversation in a direction that we can offer to the customer (more on this later).

## Models have different usability in different domains. Be creative!

A computer vision model can classify an audio signal if the imputs are spectrograms. Or you can classify (of even generate) protien sequences with an NLP model beacuse protien sequences look a lot like the inputs to an NLP model.

## The Drivetran Method

The drivetran method is a method to structure the problem you are trying to solve with AI/ML. 

1) Define the **objective** (what is the end outcome of the system?)
2) Define the **levers** (what inputs can you control?)
3) Define the **data** that needs to be captured (and do you have access to the data that relates to the objective?)
4) Define the models (how do the different models affect the outcome?) Can you complete the outcome with a model that uses less compute resources? Easier to train? Faster Inference time? Deep learning is not always the answer. Can you use a Desicion tree? An SVM? A boosted gradient machine?

```python
import time
bear_types = ["black", "grizzly", "teddy"]
path = Path("bears")
for bear in bear_types:
        time.sleep(1)
        dest = (path/bear)
        results = sum([search_google_images(f"{bear} bear", start=n) for n in range(0, 150, 10)], [])
        for idx, r in enumerate(results):
            p = dest/f"{idx}.jpg"
            download_url(r, p)
            print(f"Downloaded image to {p}")
```

```python
# create the data block
bears = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = bears.dataloaders(path)
```

```python
dls.valid.show_batch()
```

```python
learner = vision_learner(dls, resnet18, metrics=error_rate)
learner.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learner)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(10, nrows=2)
```
